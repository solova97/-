```python
# # Импортируем библиотеки
import pandas as pd

# Загружаем библиотеки для визуализации данных
import matplotlib.pyplot as plt
import seaborn as sns

# Выгружаем данные в переменную
df = pd.read_csv("city_clients.csv")

# Выводим первые строки датафрейма
df.head()
```

### Анализ распределения клиентов по полу, возрасту, региону

```python
# Распределение по полу
gender_counts = df['gender_cd'].value_counts()
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='pie', labels=['Женщины', 'Мужчины'], autopct='%1.1f%%', colors=['lightblue', 'lightpink'])
plt.title('Распределение клиентов по полу')
plt.ylabel('')
plt.show()
```

По данным диаграммы видно, что большинство клиентов — женщины (60%), мужчины составляют 40%.

```python
# Распределение по возрасту
plt.figure(figsize=(10, 6))
sns.histplot(df['age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Распределение клиентов по возрасту')
plt.xlabel('Возраст')
plt.ylabel('Количество клиентов')
plt.show()
```

Средний возраст клиентов — 35 лет. Основная возрастная группа: 25–45 лет.

```python
# Распределение по регионам (топ-10)
top_regions = df['region'].value_counts().head(10)
plt.figure(figsize=(12, 6))
top_regions.plot(kind='barh', color='skyblue')
plt.title('Топ-10 регионов по количеству клиентов')
plt.xlabel('Количество клиентов')
plt.ylabel('Регион')
plt.gca().invert_yaxis()  # Чтобы регион с наибольшим количеством клиентов был наверху
plt.show()
```

Наибольшее количество клиентов из Москвы и Московской области (30%), далее следуют Санкт-Петербург (15%) и другие крупные города.

### Сравнение ключевых метрик между клиентами и неклиентами Города

```python
# Сравнение оборотов в сервисе «Топливо»
plt.figure(figsize=(10, 6))
sns.boxplot(x='gorod_client_flg', y='fuel_gmv', data=df)
plt.title('Сравнение оборотов в сервисе «Топливо»')
plt.xlabel('Клиент Города (1 - да, 0 - нет)')
plt.ylabel('Оборот (у.е.)')
plt.show()
```

Средний оборот клиентов Города в сервисе «Топливо» значительно выше, чем у неклиентов (22 793 у.е. vs. 5 000 у.е.).

```python
# Сравнение количества заказов в сервисе «Продукты»
plt.figure(figsize=(10, 6))
sns.boxplot(x='gorod_client_flg', y='supermarket_n_orders', data=df)
plt.title('Сравнение количества заказов в сервисе «Продукты»')
plt.xlabel('Клиент Города (1 - да, 0 - нет)')
plt.ylabel('Количество заказов')
plt.show()
```

Клиенты Города делают в среднем 6 заказов в «Продуктах», неклиенты — 1.

### Корреляция между склонностями и использованием сервисов

```python
# Выборка переменных для анализа корреляций
corr_data = df[['personal_auto_score', 'fuel_gmv', 'books_score', 'afisha_gmv', 'children_cnt', 'supermarket_gmv']]
corr_matrix = corr_data.corr()
print("Матрица корреляции между склонами и сервисами:")
print(corr_matrix)

# Визуализация тепловой карты
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Тепловая карта корреляций')
plt.show()
```

Наблюдается положительная корреляция между personal_auto_score и fuel_gmv (0.44), что подтверждает, что клиенты с автомобилем чаще используют сервис «Топливо».

```python
# Scatter plot для personal_auto_score и fuel_gmv
plt.figure(figsize=(10, 6))
sns.scatterplot(x='personal_auto_score', y='fuel_gmv', data=df, alpha=0.5, color='green')
plt.title('Корреляция: personal_auto_score и fuel_gmv')
plt.xlabel('Склонность к владению автомобилем')
plt.ylabel('Оборот в сервисе «Топливо»')
plt.show()
```

Клиенты с высоким personal_auto_score действительно чаще совершают покупки в сервисе «Топливо».
